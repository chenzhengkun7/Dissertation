```python
from ML_Train.DBQuerier import DBQuerier as Querier
import numpy as np
import ast
import pandas as pd
from crawlab_toolbox import plotting
import datetime
# from datetime import datetime
# from datetime import fromtimestamp
from crawlab_toolbox.plotting import timedelta_helper
import datetime
import matplotlib.colors as colors
```

```python
fft1 = np.genfromtxt('data/bearing1FFT.csv',delimiter=',')
fft2 = np.genfromtxt('data/bearing2FFT.csv',delimiter=',')
fft3 = np.genfromtxt('data/bearing3FFT.csv',delimiter=',')
fft4 = np.genfromtxt('data/bearing4FFT.csv',delimiter=',')
times = pd.read_csv('data/bearingFFT_datetime.csv')
times = pd.to_datetime(times.values[:,1]).values
```

```python
ts = (times - np.datetime64('1970-01-01T00:00:00Z')) / np.timedelta64(1, 's')
times = [datetime.date.fromtimestamp(ts[i]) for i in range(len(ts))]
```

```python
plotting.plot_spectrogram(
            times,
            fft1,
            1,
            size=(12,4),
            showplot = True,
            save_plot = False,
            transparent = False,
            folder = None,
            filename = 'Spectrogram',
            xlabelpad = 5,    
            ylabelpad = 5,   
            file_type='pdf',
            date_format='%I:00 %p, %b %d'
            )
```
