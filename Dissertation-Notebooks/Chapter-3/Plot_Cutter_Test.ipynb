```python
import seaborn as sns
sns.set(style="ticks")

import numpy as np
from matplotlib import pyplot as plt
from crawlab_toolbox import plotting as genplt
```

```python
responses = np.load('data/Emco_Nominal_vs_Anomalous_FFT.npy',allow_pickle=True,)
frequencies = np.load('data/frequencies.npy')
nominal_power = np.load('data/nominal_power.npy')
anomalous_power = np.load('data/anomalous_power.npy')
nominal_rms = np.load('data/nominal_rms.npy')
anomalous_rms = np.load('data/anomalous_rms.npy')
```

```python

labels = [r'Nominal','Anomalous']

# Create a 2D plot
genplt.generate_plot(frequencies/1000, # X - coordinate on the plot
              responses,# Y - coordinates on the plot
              labels,# Labels for the plot legend
              'Frequency (kHz)',# X - axis label
              r'PSD ($V^2 / Hz$)',   # Y - axis label
              filename='Emco_Nominal_vs_Anomalous_FFT',   # Plot filename
              folder='Figures',    # Specify the folder where the filename will be saved
              num_col=1,    # Specify the number of columns in the legend
              legend_loc='best', # Specify the location of the legend
              ymax=0.1,
                     log_y=False,
                     ymin=0.05,
              transparent=False,
              showplot=True,
              save_plot=True,
              template='wide',
              file_type='png'
             )
```

```python
num_col = 2
legend_loc = 'upper right'
transparent = False

plt.figure(figsize=(12,4))
ax = plt.gca()
# the histogram of the data
n, bins, patches = plt.hist(nominal_power, 10, density=True, facecolor='b', alpha=0.50,label='Nominal')
n, bins, patches = plt.hist(anomalous_power, 10, density=True, facecolor='r', alpha=0.50,label='Anomalous')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xlabel(r'\textbf{Spindle Power (W)}')
plt.ylabel(r'\textbf{Probability}')
ax.legend(ncol=num_col,loc=legend_loc,framealpha=float(not transparent)).get_frame().set_edgecolor('k')
plt.grid(False)
plt.savefig('Figures/Spindle_Load_Anomalous_Compare.png',transparent=True)     
plt.show()
```

```python
num_col = 2
legend_loc = 'upper right'
transparent = False

plt.figure(figsize=(12,4))
ax = plt.gca()
# the histogram of the data
n, bins, patches = plt.hist(nominal_rms, 5, density=True, facecolor='b', alpha=0.50,label='Nominal')
n, bins, patches = plt.hist(anomalous_rms, 5, density=True, facecolor='r', alpha=0.50,label='Anomalous')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xlabel(r'\textbf{Vibration RMS}')
plt.ylabel(r'\textbf{Probability}')
ax.legend(ncol=num_col,loc=legend_loc,framealpha=float(not transparent)).get_frame().set_edgecolor('k')
plt.grid(False)
plt.savefig('Figures/RMS_Anomalous_Compare.png',transparent=True)  
plt.show()
```
